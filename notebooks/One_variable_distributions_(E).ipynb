**Task:**

1. Load train.csv with pandas.

2. For Age and Fare, compute mean, median, mode, variance, standard deviation, and
interquartile range.

3. Plot histogram and boxplot for each and comment on skewness using sample skewness
formula and visual intuition.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"  # Titanic
df = pd.read_csv(url)
#df
```

```python
Age_mean = df["Age"].dropna().mean()
Age_median = df["Age"].dropna().median()
Age_mode = df["Age"].dropna().mode()
Age_variance = df["Age"].dropna().var()
Age_std = df["Age"].dropna().std()
q1 = df["Age"].dropna().quantile(0.25)
q3 = df["Age"].dropna().quantile(0.75)
IQR = q3-q1
```

```python
print('Mean',Age_mean)
print('Median',Age_median)
print('Mode',Age_mode)
print('Variance',Age_variance)
print('Standard deviation',Age_std)
print('Inter-Quantile range',IQR)
```

```python
plt.figure(figsize=(8,3.5))
plt.subplot(1,2,1)
sns.histplot(df["Age"].dropna(),bins=30)
plt.title("Histogram of Age")

plt.subplot(1,2,2)
sns.boxplot(df["Age"].dropna())
plt.title("Box Plot of Age")
plt.show()
```
